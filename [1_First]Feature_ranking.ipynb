```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import RidgeCV, LinearRegression
from sklearn import metrics
import sys
import copy
import statsmodels.api as sm
pd.options.mode.chained_assignment = None
from statsmodels.stats.outliers_influence import variance_inflation_factor
pd.options.display.max_rows = 4000
pd.options.display.max_seq_items = 2000
%matplotlib inline

dfTemp=pd.read_csv(r'C:\Users\Jaesung Park\Desktop\FP\GroupProject\final.csv',keep_default_na=False)
```

```python
dataset = pd.get_dummies(dfTemp, drop_first=True)
dataset = dataset[dataset.columns.difference(['PID','Order','Total Bsmt SF'])]
# mix the dataset
dataset = dataset.reindex(np.random.permutation(dataset.index))

all_columns = [] 
for i in range(len(dataset.columns)):
    if dataset.columns[i] not in (['nothing']):
        all_columns.append(dataset.columns[i])
```

```python
regressor = LinearRegression(fit_intercept=True)

#alphas = [0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1,0.5, 1, 10, 15, 20, 30, 40, 100]
#regressor=RidgeCV(alphas=alphas, store_cv_values=True)
```

```python
alphasL = []
r2_list = []
predicted = []
columns_name = []
```

```python
second = dataset[dataset.columns.difference(['Saleprice'])]
```

```python
for j in range(len(second.columns)):
    mylist = copy.deepcopy(dataset)
    appended_data = []
    
    
    x_data = second[second.columns[j]].values.reshape(-1,1)
    y_data = dataset['SalePrice']
    x_train,x_test,y_train,y_test = train_test_split(x_data,y_data)
  
    testing_set_X = x_test
    testing_set_Y = y_test
    training_set_X = x_train
    training_set_Y = y_train

    ridge_mod = regressor.fit(training_set_X, training_set_Y)
  #  alphasL.append(ridge_mod.alpha_)
    y_pred = regressor.predict(testing_set_X)
    r2 = metrics.r2_score(testing_set_Y, y_pred)
    py = y_pred
    
    columns_name.append(dataset.columns[j])
    r2_list.append(r2)
    predicted.append(py)
```
